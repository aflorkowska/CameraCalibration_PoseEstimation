W większości przypadków każda kamera (a dokładniej obiektyw i soczewka) wprowadza jakieś zniekształcenia do obrazu (głównie dotyczy to soczewek z szerokim polem widzenia, ekstremalny przypadek to soczewki rybiego oka (fisheye lens)). Innymi słowy, zarejestrowany obraz może niedokładnie odzwierciedlać rzeczywistość. Przykładowo, prosta linia w rzeczywistości, jest zakrzywiona na obrazie lub przesunięta. Wynika to z istniejących dwóch głównych typów zniekształceń: 

- radialnych (radial distortion) 
    - powód: wynikają ze kształtu soczewki, co prowadzi do różnego stopnia załamania światła w różnych częściach soczewki, co powoduje nieliniowy przebieg promieni świetlnych przez soczewkę
    - efekt wizualny: powoduja, że proste linie w rzeczywistości są zakrzywione na obrazie. Kształt soczewki powoduje, że powiększenie obrazu zmniejsza się lub zwiększa promieniowo od środka obrazu. Wyróżniamy dwa podtypy:
        - beczkowych (barrel distortion) - zakrzywienie do zewnątrz 
        - poduszkowatych (pincushion distortion) - zakrzywienie do wewnątrz
- tangensialnych (tangential distortion).
    - powód: niedoskonałość montażu optyki soczewki - soczewka i płaszczyzna obrazu nie są równoległe, innymi słowy oś optyczna nie jest prostopadła do kamery.
    - efekt wizualny: proste linie są zniekształcone w kierunku, który jest równoległy do osi soczewki.

Co więcej, każda kamera ma swoje uniwersalne zniekształcenia, co oznacza, że każdą kamerę należy analizować osobno, jako oddzielny przypadek. Takie zniekształcenia można zniwelować dzięki kalibracji kamery i określeniu współczynników zniekształcenia.
