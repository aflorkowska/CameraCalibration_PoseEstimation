### Słowem wstępu
Czasami w wizji komputerowej używa się wartości wektorów i wartości własnych.

Jeśli: 
    Macierz A transformuje przestrzeń (np. operacja mnożenia), ale niektóre kierunki pozostają bez zmian – jedynie ich długość się zmienia. Te kierunki to wektory własne, a skala tej zmiany to wartości własne.

Wektory własne (eigenvectors) wskazują kierunki, które pozostają niezmienione przez transformacje
Wartości własne (eigenvalues) określają jak bardzo macierz skaluje się wzdłuż tych kierunków.
- value > 1 : wektor zostaje rozciągnięty (wzmocniony)
- 0 < value < 1 : wektor zostaje skompresowany (osłabiony)
- value = 1 : długość się nie zmienia
- value = 0 : wzdłuż tego kierunku nie ma żadnej zmienności danych, czyli punkty leżą w podprzestrzeni niższego wymiaru, więc można go zredukować. np. jeśli macierz transformacji ma wartość własną 0 w danym kierunku, to wszystkie punkty są ściskane do punktu lub płaszczyzny. W analizie obrazów z kolei: jeśli transformacja daje wartość własną 0, to pewien kierunke nie zawiera żadnej informacji np. utrata kontrastu w danym kierunku.
- value < 0 : następuje zmiana zwrotu wektora własnego (obrót o 180)

### Singular Value Decomposition

To znana i użyteczna technika przy działaniach na macierzach, która bierze macierz M (m x n) i rozkłąda na czynniki do 3 macierzy. I te macierze mówią o strukturze i właściwościach tej macierzy pierwotnej M. Jest uogólnieniem dekopompozycji wartości własnych (Eigenvalue) dla macierzy niekwadratowych:
- U - macierz lewych wektorów singularnych (m x m) - zawiera ortogonalne wektory własne dla AA^T (baza wierszy macierzy A). Wektory w U sa bazą dla przestrzeni wynikowej - opisują kierunek po transformacji.
    - użyteczne np. przy kompresji obrazów np. do JPG, kiedy wybieramy wiersze macierzy U które przedstawiają główne struktury w obrazie (macierz wejściowa A)
    - jeśli macierz A zawieraj informacje o obserwacjach punktów 2D z różnych kątów, to jest to użyteczne przy structure from motion albo rekonstrukcji 3D, gdyż wiersze U zawierają projekcie punktów na główne osie wariancji i mogą reprezentować kształt obiektu lub jego orientację.
- D to macierz diagonalna, która przechowuje singularne wartości (singular values) macierzy M, które mówią o sile wpływu danej składowej. Uporządkowane  malejąco, co oznacza, że pierwsze wartości reprezentują najważniejsze informacje.
- V^T - macierz prawych wektorów singularnych (n x n) - zawiera ortogonalne wektory własne dla A^TA (baza wierszy macierzy A). Wektory w V są bazą przestrzeni wejściowej (kolumn A), a V^T są bazą przestrzenii wejściowej tyle, że wierszy A - opisują jak dane wejściowe są przekształcane przez A.


Jest przydatne przy rozwiązywaniu wielu problemów matematycznych np. przy liniowych równaniu współrzędnych homogenicznych (przedstawione niżej), lub np. w PCA, gdzie obliczamy macierz kowariancji i obliczamy wartości i wektory własne. Wartości własne określają jak dużo informacji zawiera dany kierunek, a wektor mówi o kieunku największej zmienności. Dzięki temu możemy wybrać te wektory własne, które zawierają wartość informacji powyżej jakiegoś ustalonego progu (zazwyczaj suma wartości > = 95% wariancji), tym samym redukując złożoność.
