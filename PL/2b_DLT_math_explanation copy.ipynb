### Direct Linear Transformation - oblicza macierze K, R i T ale punkty kontrolne NIE MOGĄ należeć do jednej płaszczyzny.
(Polecam obejrzeć wyjaśnienie Cyrila Stachnissa: https://www.youtube.com/watch?v=3NcQbZu6xt8)

Dane:
- co chcemy obliczyć? Parametry wewnętrzne i zewnętrzne kamery
- co jest dane: współrzędne 3D punktów kontrolych (narożniki szachownicy) oaz współrzędne 2D tych punktów na obrazie


Direct Linear Transformation (DLT) - mapuje jakikowiek punkt 3D do punktu 2D na płaszczyznie obrazu. Jeśli zakładamy idealną soczewkę (bez zniekształceń) i tylko 11 nieznanych parametrów, to może być obliczona przy min. 6 punktach kontrolnych. Co ważne, punkty wejściowe NIE MOGĄ LEŻEĆ NA JEDNEJ PŁASZCZYŹNIE więc wzornik szachownicy odpada !!! Jeśli będą leżeć na jednej płaszczyznie to przy min 6 punktach z 12 równań robi się 9 < 11 -> nie jesteśmy w stanie rozwiązać go dla 11 niewiadomych. Również jeśli wykryte punkty leżą na obiekcie umieszczonym daleko od nas (np. ziemia z widoku z samolotu) to następuję przybliżenie, że te punkty leżą na jednej płaszczyźnie.
