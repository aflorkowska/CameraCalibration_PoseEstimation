### Stereowizja
Podczas akwizycji obrazu przez kamery następuje proces projekcji perspektywy. Polega on na przekonwertowaniu punktów 3D na płaszczyznę 2D, tym samym powoduje utracenie informacji o głębi czyli o odległości każdego punktu (piksela) od obiektu aparatu. Jednakże zastosowanie układu stereo kamer pozwala na rekonstrukcję 3D punktów bazując na obrazach 2D. Składa się on z dwóch kamer umieszczonych w znanej odległości i wykonujących zdjęcia w tym samym czasie (na tej samej wysokości na osi Y i Z). Najprostszym przykładem takiego układu są ludzkie oczy. Przy opisie sceny obserwowanej z wykorzystaniem dwóch kamer kluczowym etapem jest zdefiniowaniem zewnętrznego układu współrzędnych, wspólnego dla obu systemów akwizycji obrazów. Z reguły nowy układ współrzędnych definiowany jest zgodnie z pozycją jednej z kamer, co w dalszej kolejności sprowadza się do wyznaczenia transformacji sztywnej pomiędzy obydwoma urządzeniami. W efekcie, dzięki kalibracji takiego układu stereo, dostarczona jest informacja o macierzy rotacji i translacji, które wiążą poszczególne układy ze sobą. Pozwala to następnie na odniesienie do siebie danych obrazowych uzyskanych z różnych pozycji. Dodatkowo wyznaczona zostaje macierz zasadnicza E (ang. essential matrix, bazująca tylko na macierzy rotacji i translacji) i fundamentalna F  (ang.  fundamental matrix, macierz zasadnicza, która dodatkowo uwzględnia parametry wewnętrzne obu kamer).

### Procedura kalibracji dwóch kamer stereo
source: https://dibyendu-biswas.medium.com/stereo-camera-calibration-and-depth-estimation-from-stereo-images-29d87bc702f3

1. Pozyskanie zdjęć kalibracyjnych przedstawiających ten sam wzorzec, z różnych kątów i odległości. Należy zadbać, aby na każdym ujęciu była widoczna cała szachownica. 
2. Przeprowadzenie wstępnej kalibracji każdej z kamer z osobna: uzyskanie macierzy parametrów wewnętrznych i współczynnik dystorsji dla każdej z kamer z osobna.
3. Przeprowadzenie kalibracji układu stereoskopowego i uzyskanie macierzy rotacji i wektora translacji okreslającego położenie i orietację kamer względem siebie. 

### A jak określić położenie punktu w przestrzeni na podstawie danych z kamer stereo?

Można wyróżnić dwa główne podejścia:
- podejście oparte na bazie stereo (baseline) i paralaksie dedykowane dla skalibrowanych kamer stereo o równoległych osiach optycznych
- podejście oparte na macierzach rzutowania i triangulacji dedykowane dla dowolnie ustawionych kamer stereo
